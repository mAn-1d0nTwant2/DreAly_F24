```python
import numpy as np
import pandas as pd
import math
import statistics
from scipy import stats
```

```python
belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd. read_csv(belize_listings_path)
belize_df. describe ()
```

```python
grouped_belize_df = belize_df.groupby('neighbourhood')
print(grouped_belize_df)
```

```python
grouped_belize = grouped_belize_df["price"].mean()
```

```python
grouped_df = grouped_belize_df[["number_of_reviews", "availability_365"]].max()
```

```python
merged_df = pd.merge(grouped_belize, grouped_df, on = "neighbourhood")
merged_df
```
